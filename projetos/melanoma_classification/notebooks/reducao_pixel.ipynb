```python
df_train_GroundTruth = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/proyecto/ISIC_2020_Training_GroundTruth.csv')
df_train_GroundTruth.head()
```

```python
from numpy.random.mtrand import randint
from PIL import Image
import matplotlib.pyplot as plt

ruta1 = "/content/drive/MyDrive/Colab Notebooks/proyecto/ISIC_2020_Training_JPEG/train/"
for i in range(len(df_train_GroundTruth)):
  name = df_train_GroundTruth.image_name[i] + ".jpg"
  imagen = Image.open(ruta1 + name).convert('RGB')
  ancho, alto = imagen.size
  print(ancho,alto)
  nuevo_ancho = 640
  factor = nuevo_ancho / ancho
  nuevo_alto = int(alto * factor)
  imagen_resized = imagen.resize((nuevo_ancho, nuevo_alto), resample=Image.BILINEAR)
  ruta2 = "/content/drive/MyDrive/Colab Notebooks/proyecto/ISIC_2020_Training_JPEG/train/target_" + str(df_train_GroundTruth.target[i]) + "/"
  imagen_resized.save(ruta2 + name)
```

```python
plt.imshow(imagen_resized)
plt.show()
plt.imshow(imagen)
plt.show()
```
