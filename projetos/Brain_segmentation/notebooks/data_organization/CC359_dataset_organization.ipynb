# Organização do conjunto de dados CC359

## Criar pasta para cada sujeito

* Essa pirmeira parte do código apenas separa os arquivos de cada sujeito em pastas, tanto as imagens que então dentro de uma pasta chamada "Original" como as segmentações que estão dentro de uma pasta chamada "Staple" são transferidas para uma pasta nomeada de "CC359" em "organized_dataset". Dentro desta pasta o código abaixo criar uma pasta para cada sujeito adicionando "CC" mais os caracteres que estão após o "CC" no nome do arquivo. O arquivo com o mesmo número que a pasta é criada é trasferido para dentro desta pasta.

```python
import os
import glob
import shutil
# Obter todos os arquivos na pasta original\n
#file_paths = glob.glob(\"/home/joany/Downloads/CC359/Original/*.nii.gz\")
file_paths = glob.glob("/home/joany/brain_dataset/raw_dataset/CC359/Staple/*.nii.gz")
for data_path in file_paths:
# Extrair o nome do arquivo
    file_name = os.path.basename(data_path)
    # Extrair o padrão 'CC' seguido do número do nome do arquivo
    pattern = 'CC'
    start_index = file_name.find(pattern)
    end_index = start_index + len(pattern) + 4  # Incluir os 4 caracteres após 'CC'
    patient_id = file_name[start_index:end_index]
    print(patient_id)
    # Criar a pasta para o sujeito (se ainda não existir)
    folder_path = "/home/joany/brain_dataset/organized_dataset/CC359/" + patient_id
    os.makedirs(folder_path, exist_ok=True)
    # Mover o arquivo para a pasta do sujeito
    shutil.copy(data_path, folder_path)
```

# Mudar o nome dos arquivos 

* Essa segunda parate apenas troca o nome do arquivo para padrozizar em todos as pastas. Para cada arquivos foi usado o padrão contido no nome do arquivo. Por exemplo, as imagens tem os padrões de "M.nii.gz" ou "F.nii.gz" (foi preeciso executar o código duas vezes trocando o "M" por "F") e as segmentações tem o padrão de "staple" então foi preciso executar uma única vez. Os nomes das imagens passou a ser 'brain_T1w.nii.gz' e as das segmentações de 'brainmask_T1w.nii.gz'. Note que esses nomes são usados para todos os conjuntos de dados, para facilitar as execuções futuras.

```python
import os
import glob
def padronizar_nomes_pasta(caminho_pasta):
    subjects = glob.glob(os.path.join(caminho_pasta, "*"))
    for subject in subjects:
        if os.path.isdir(subject):
            print(subject)
            arquivos = glob.glob(os.path.join(subject, '*_staple.nii.gz'))
            for caminho_completo in arquivos:
                if os.path.isfile(caminho_completo):
                    novo_nome = 'brainmask_T1w.nii.gz'
                    novo_caminho = os.path.join(subject, novo_nome)
                    os.rename(caminho_completo, novo_caminho)
                    print(f"Arquivo {os.path.basename(caminho_completo)} foi renomeado para {novo_nome}.")

pasta = "/home/joany/brain_dataset/organized_dataset/CC359/"
padronizar_nomes_pasta(pasta)
```

## separar dados em treinamento e validação

* O código abaixo separa o conjunto de dados CC359 em treinamento (80%) e validação (20%). Antes de realizar a execução deste código os 12 sujeitos que possui máscaras manuais foram separados manualmente. O resultado da divisão dos dados é salvo em "split_dataset/CC359".

```python
import os
import random
import shutil
from sklearn.model_selection import train_test_split
# Definir caminho para a pasta original com os dados
pasta_origem = "/home/joany/brain_dataset/organized_dataset/CC359"
# Definir caminho para a pasta de destino
pasta_destino = "/home/joany/brain_dataset/split_dataset/CC359"

# Criar as pastas de treinamento, validação e teste
pasta_treinamento = os.path.join(pasta_destino, "train")
pasta_validacao = os.path.join(pasta_destino,"val")
os.makedirs(pasta_treinamento, exist_ok=True)
os.makedirs(pasta_validacao, exist_ok=True)
# Lista para armazenar os nomes das pastas
nomes_pastas = []
# Percorrer todas as pastas na pasta original
for nome_pasta in os.listdir(pasta_origem):
    caminho_pasta = os.path.join(pasta_origem, nome_pasta)
    if os.path.isdir(caminho_pasta):
        nomes_pastas.append(nome_pasta)

# Dividir os nomes das pastas em treinamento e validação
nomes_treinamento, nomes_validacao = train_test_split(nomes_pastas, train_size=0.8, random_state=42)

# Mover as pastas de treinamento para a pasta de treinamento
for nome_pasta in nomes_treinamento:
    origem_pasta = os.path.join(pasta_origem, nome_pasta)
    destino_pasta = os.path.join(pasta_treinamento, nome_pasta)
    shutil.move(origem_pasta, destino_pasta)

# Mover as pastas de validação para a pasta de validação
for nome_pasta in nomes_validacao:
    origem_pasta = os.path.join(pasta_origem, nome_pasta)
    destino_pasta = os.path.join(pasta_validacao, nome_pasta)
    shutil.move(origem_pasta, destino_pasta)
```
