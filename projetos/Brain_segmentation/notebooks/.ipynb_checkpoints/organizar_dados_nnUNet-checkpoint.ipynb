## 1.2. Renomear arquivos
Usei para cada pasta diferente, nomes identificados como padrão para os nomes dos arquivos. Ex.: Para as imagens do dataset NFBS usei (*_T1w.nii.gz) e para as segmentações usei (*_T1w_brainmask.nii.gz). Para as imagens do dataset CC359 usei (*_F.nii.gz ou *_M.nii.gz) e para as segmentações usei (*_staple.nii.gz). Para as imagens do dataset LBPA40 usei (*native.mri.nii.gz) e para as segmentações usei (*native.brain.mask.nii.gz). Todas as imagens de todos os datasets foram renomedos de "brain_T1w.nii.gz" e todas as segmentações de todos os datasets foram renomeados de 'brainmask_T1w.nii.gz'. 

```python
import os
import glob

def padronizar_nomes_pasta(caminho_pasta):
    
    subjects = glob.glob(os.path.join(caminho_pasta, "*"))
    
    for subject in subjects:
        if os.path.isdir(subject):
            print(subject)
            arquivos = glob.glob(os.path.join(subject, '*native.brain.mask.nii.gz'))
            
            for caminho_completo in arquivos:
                if os.path.isfile(caminho_completo):
                    novo_nome = 'brainmask_T1w.nii.gz'
                    novo_caminho = os.path.join(subject, novo_nome)
                    os.rename(caminho_completo, novo_caminho)
                    print(f"Arquivo {os.path.basename(caminho_completo)} foi renomeado para {novo_nome}.")

# Exemplo de uso
pasta = "/home/joany/Brain_projeto_IA901A"
padronizar_nomes_pasta(pasta)
```

Esses processos foram feitos cuidadosamente em cada conjunto de dados. 

# 2. Organização de dados para o treinamento do framework nnU-Net
É preciso organizar os dados da maneira que a nnU-Net exigi. Para isso foi feito a seguinte implementação, que organizou os dados de acordo com o descrito em: 

Obsrvação: para a organização dos dados como a nnU-net exige, é preciso realizar as configurações de framework. 

```python
import os 
import glob 
import shutil

for image_path in glob.iglob("bkp_Brain_projeto_IA901A/train/**/brain_T1w.nii.gz", recursive=True):
    ID = os.path.basename(os.path.dirname(image_path))

    dst_path = os.path.join("imagesTr", f"{ID}_0000.nii.gz")
    print(f"Copying {image_path} -> {dst_path}")

    shutil.copy2(image_path, dst_path)

for mask_path in glob.iglob("bkp_Brain_projeto_IA901A/train/**/brainmask_T1w.nii.gz", recursive=True):
    ID = os.path.basename(os.path.dirname(mask_path))

    dst_path = os.path.join("labelsTr", f"{ID}.nii.gz")
    print(f"Copying {mask_path} -> {dst_path}")

    shutil.copy2(mask_path, dst_path)

for test_image_path in glob.iglob("bkp_Brain_projeto_IA901A/test/**/brain_T1w.nii.gz", recursive=True):
    ID = os.path.basename(os.path.dirname(test_image_path))

    dst_path = os.path.join("imagesTs", f"{ID}_0000.nii.gz")
    print(f"Copying {test_image_path} -> {dst_path}")

    shutil.copy2(test_image_path, dst_path)

for test_mask_path in glob.iglob("bkp_Brain_projeto_IA901A/test/**/brainmask_T1w.nii.gz", recursive=True):
    ID = os.path.basename(os.path.dirname(test_mask_path))

    dst_path = os.path.join("labelsTs", f"{ID}.nii.gz")
    print(f"Copying {test_mask_path} -> {dst_path}")

    shutil.copy2(test_mask_path, dst_path)
```
